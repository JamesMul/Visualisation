```python
# Barchart plot of 4 main effects of 


N = 4
score_pre = (70.6277,33.8564,69.6586,28.6818) #average score (same order as spss is)
prestd = (11.21693,11.71888,11.18791,15.43886) #standard deviation
fig, ax = plt.subplots()

ind = np.arange(N)    # the x locations for the groups
width = 0.35         # the width of the bars
p1 = ax.bar(ind, score_pre, width, bottom=0, color="lightsteelblue", yerr = prestd, capsize=5, linewidth=2) #customise bar plot


score_post = (70.1391,37.7773,70.8432,32.7668)
poststd = (13.68356,12.14094,12.44354,15.49278)

p2 = ax.bar(ind + width, score_post, width, bottom=0, yerr = poststd, capsize = 5)

#####ax.set_title('Mean intensity rating across cardiac and emotion axis', fontsize=14)
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('Diastole fear', 'Diastole neutral', 'Systole fear', 'Systole neutral'))
plt.xticks(rotation=30, fontsize = 14)
ax.legend((p1[0], p2[0]), ('Pre-training', 'Post-training'), loc='lower left', bbox_to_anchor=(1, 0.5), frameon = False, fontsize = 14)
ax.autoscale_view()

#removing top and right borders
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Hide grid lines
ax.grid(False)

#set y range
plt.ylim([0, 100])


#change interval
import matplotlib.ticker as plticker
loc = plticker.MultipleLocator(base=15.0) # this locator puts ticks at regular intervals
ax.yaxis.set_major_locator(loc)

#change y axis font size
ax.tick_params(axis="y", labelsize=14)
ax.spines['left'].set_linewidth(1)
ax.spines['bottom'].set_linewidth(1)
ax.spines['left'].set_color('black')
ax.spines['bottom'].set_color('black')

plt.ylabel('Intensity rating', fontsize = 14)

# plt.gcf()
# #only use when saving figure
# os.chdir('file path')
# plt.savefig('summary_fig1.png', format='png', dpi=1000, bbox_inches='tight')

# plt.show()
```
